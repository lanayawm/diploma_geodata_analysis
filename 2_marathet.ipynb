```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy import stats
```

```python
file = pd.read_excel(')
# (oc - mq)/ mq | (foc - mq)/mq
data = file['FOC-MQ']
# data = file['OC']
first = min(data)
last = max(data)

nodes = len(data)
disperssion = round(np.var(data), 2)
std = round(np.std(data), 2)
mean = round(np.mean(data), 2)
median = round(np.median(data), 2)
print(f"{round(std, 2)}, {round(mean, 2)}, {round(median, 2)}")
```

```python
# bins = np.logspace(np.log10(min(data)), np.log10(max(data)), num=300)
plt.hist(data, bins=20, edgecolor='black', color='green', density=True)
# plt.yscale('log')
# plt.xscale('log')
plt.xlabel('Значение отклонения в %')
plt.ylabel('Частота')
plt.title('Распределение значений отклонений Focused от Marquardt')
# plt.xticks([1, 10, 30, 50, 100, 300, 1000, 3000], labels=['1', '10', '30', '50', '100', '300', '1000', '3000'])

norm_x = np.linspace(first, last, 100)
plt.plot(norm_x, stats.norm.pdf(norm_x, mean, std))

x_min, x_max = plt.gca().get_xlim()
y_min, y_max = plt.gca().get_ylim()
plt.text(x_max * 0.3, y_max * 0.75, f"D = {disperssion}\nσ = {std}%\nСр.знач = {mean}%\nМедиана = {median}%\nКол-во точек = {nodes}")

plt.show()
```
