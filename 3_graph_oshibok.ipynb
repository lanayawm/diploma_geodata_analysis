```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# пытаюсь задать все переменные
pickets = [100, 600, 900, 1230, 1700, 2000]
file = pd.read_excel('D:/Learning/pythonich/Работа_с_библиотеками/for_diploma/Usual_razn.xlsx', sheet_name='2750_разность')
values = []
y = file[file['X'] == 0]['Y']
```

```python
for picket in pickets:
    data = file[file['X'] == picket]['razn_foc']
    values.append(data)

# print(len(values))
```

```python
# много графиков
fig, axs = plt.subplots(len(pickets) // 2, 2, figsize=(15, 10))
axs = axs.flatten()
for i in range(len(pickets)):
    axs[i].plot(values[i], y, color='blue')
    axs[i].set_title(f'Пикет - {pickets[i]}')
    axs[i].set_xlabel(f'Отклонение %')
    axs[i].set_ylabel(f'Глубина, м')
    axs[i].set_xlim(-10, 45)
    axs[i].grid()

fig.suptitle('Графики ошибок по пикетам', fontsize=14)
plt.tight_layout() # так графики не накладываются на друг друга
plt.show()
```

```python
# один график
plt.figure(figsize=(8, 5))
for i in range(len(values)):
    plt.plot(values[i], y, label=f'Пикет - {pickets[i]}')  # Строим каждую линию из 'values' против 'y'

plt.title('Графики ошибок по пикетам')
plt.xlabel('Отклонение %')
plt.ylabel('Глубина, м')
plt.legend()
plt.grid()
plt.show()
```
